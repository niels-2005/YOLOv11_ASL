```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")
# Dataset File
DATA = "data.yaml"

# Tuned Hyperparameters, added Early Stopping
results = model.train(
    data=DATA, 
    epochs=500, 
    patience=50, 
    lr0=0.00581,
    lrf=0.00679,
    momentum=0.91582,
    weight_decay=0.00048,
    warmup_epochs=3.79318,
    warmup_momentum=0.95,
    box=5.17318,
    cls=0.49701,
    dfl=1.44457,
    hsv_h=0.01621,
    hsv_s=0.57263,
    hsv_v=0.45212,
    degrees=0.0,
    translate=0.12153,
    scale=0.41088,
    shear=0.0,
    perspective=0.0,
    flipud=0.0,
    fliplr=0.53931,
    mosaic=0.85179,
    mixup=0.0,
    copy_paste=0.0,
    optimizer="AdamW"
)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os 

train_path = "./runs/detect/train52/"
confusion_matrix_path = os.path.join(train_path, "confusion_matrix.png")
results_path = os.path.join(train_path, "results.png")
val_batch_path = os.path.join(train_path, "val_batch0_labels.jpg")

image_paths = [confusion_matrix_path, results_path, val_batch_path]

for image_path in image_paths:
    plt.figure(figsize=(12, 12))
    img = mpimg.imread(image_path)
    plt.imshow(img)
    plt.axis('off')  # Hide the axis
    plt.show()
```
